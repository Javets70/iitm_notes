### 1) Write a function compute_GridSearchCV which accepts the Kernel and regularization parameters as inputs and returns the Mean cross-validated score of the best_estimator, denoted with best_score_ of the models with the below-mentioned hyperparameters:

Split the Iris dataset into train and test sets with 70:30 ratio

Import svm.SVC as 'model'

```python
from sklearn.datasets import load_iris 

X,y = load_iris(return_X_y = True , as_frame=True)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test , y_train , y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.model_selection import GridSearchCV

def compute_GridSearchCV(model , params , X , y , cv):
    gridSearch = GridSearchCV(model , params , cv=cv)
    gridSearch.fit(X,y)
    return gridSearch.best_score_
```

```python
from sklearn.svm import SVC

params = {
    "kernel" : ["linear" , "rbf"],
    "gamma" : ['auto'],
    "C" : [1,15,25]
}
compute_GridSearchCV(SVC(random_state=0) , params , X_train , y_train , 4)
```

## Common Data for 2 to 3 questions
Split the Social_Network_Ads dataset
(https://drive.google.com/file/d/1qUa1GlG4X4ZY_4E0e7jPR-z7AG7NIDbE/view?usp=sharing) into training and test set in 75:25 ratio.  
Fit and transform the train and test set of the feature matrix by applying StandardScaler transformer.  
Fit a linear SVM (with random_state = 0and linear kernel) to training data.  

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["Purchased"])
y = df.Purchased

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25 , random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

clf = SVC(random_state=0 , kernel='linear')
clf.fit(X_train , y_train)
```

### 2) The predicted data returns an accuracy_score on test data. Which of the following option represents the calculated accuracy_score?

```python
clf.score(X_test , y_test)
```

### 3) Calculate the confusion matrix obtained from the above-predicted data.

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test , clf.predict(X_test))
```

**From the MNIST dataset, consider the first 20,000 data samples as training data and the next 5,000 data samples as test data. Fit a pipeline with MinMaxScaler and a classifier with SVC, linear kernel, one vs rest decision_function_shape and class_weight=None to this dataset and answer the following questions(Q.no 4 and Q.no 5).**

```python
from sklearn.datasets import fetch_openml

df = fetch_openml("mnist_784")
```

```python
X = df['data']
y = df['target']
```

```python
X_train = X[:20000]
X_test = X[20000:25000]

y_train = y[:20000]
y_test = y[20000:25000]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ("scaling" , MinMaxScaler()),
    ("clf" , SVC(kernel='linear' , class_weight=None , decision_function_shape='ovr')),
])
pipe.fit(X_train , y_train)
```

### 4) What is the sum of the main diagonal elements of the confusion matrix?

```python
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test , pipe.predict(X_test))
```

```python
matrix.trace()
```

### 5) Which of the following are the correct values of precision, recall and f1_Score?

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , pipe.predict(X_test)))
```
